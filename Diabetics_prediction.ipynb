```python
from sklearn.datasets import load_diabetes
```

```python
dataset=load_diabetes()
```

```python
dataset
```

```python
print(dataset["DESCR"])
```

```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
X=pd.DataFrame(dataset.data,columns=["age","sex","bp","s1","s2","s3","s4","s5","s6","s7"
])
y=dataset['target']
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(X,y,test_size=0.25,random_state=10
)
```

```python
x_train.corr()
```

```python
import seaborn as sns 
plt.figure(figsize=(15,10))
sns.heatmap(x_train.corr(),annot=True)
```

```python
from sklearn.tree import DecisionTreeRegressor
regressor =DecisionTreeRegressor()
regressor.fit(x_train,y_train)
```

Hyper parameter tuning

```python
param ={ 
    'criterion': ['squred_error',"friedman_mse","absolute_error"],
    "splitter":["best","random"],
    "max_depth":[1,2,3,4,5,10,15,20,25],
    'max_features':["auto","sqrt","log2"]
}
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
regressor=DecisionTreeRegressor()
```

```python
grid=GridSearchCV(regressor,param_grid=param,cv=5,scoring="neg_mean_squared_error")
```

```python
import warnings
warnings.filterwarnings("ignore")
grid.fit(x_train,y_train)
```

```python
grid.best_params_
```

```python
y_pred=grid.predict(x_test)
```

```python
from sklearn.metrics import r2_score,mean_squared_error,mean_squared_error,mean_absolute_error
```

```python
print(r2_score(y_test,y_pred))
print(mean_absolute_error(y_test,y_pred))
print(mean_squared_error(y_test,y_pred))
```

```python
select_model.fit(x_train,y_train)
```
